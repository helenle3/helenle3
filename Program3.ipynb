```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Loading into pandas dataframe
fire_area = pd.read_csv('forestfires.csv')
```

```python
#Graph 1: Fire Area vs Temperature

#Construct Linear Regression
temp_linregress = stats.linregress(x=fire_area.Temperature,
                                     y=fire_area.Fire_Area)

#Print Out Assessment Statistics
print("Graph 1: ")
print(f"The slope is: {temp_linregress.slope}")
print(f"The intercept is: {temp_linregress.intercept}")
print(f"The r^2 value is {temp_linregress.rvalue ** 2}")
print(f"The p-value is {temp_linregress.pvalue}")

#Print Out Equation and Calculate Prediction
print()
print(f"The linear regression equation is y = {temp_linregress.slope} * x + {temp_linregress.intercept}")
temp_prediction = temp_linregress.slope * 23 + temp_linregress.intercept
print(f"The predicted fire area when temperature is 23 Celsius is {temp_prediction:.2f} units.")

#Analysis
print()
print("Given the r^2 value of approximately 0.0096, this means that the independent variable, temperature, can explain only 0.96% of the variation in the dependent variable, fire area. This indicates that other factors are more significant in assessing fire area. While temperature may still have some effect, the relationship is weak. 99.04% of the variation in fire area is explained by external factors not captured by temperature.")
print()
print("With a p-value of approximately 0.0261, assuming that we are at a confidence interval of 95%, since our p-value is less than the alpha of 0.05, temperature has a statistically significant relationship with fire area. This means that temperature has some effect on fire area and is not purely random.")

#Construct Graph
temp_model = sns.regplot(x=fire_area.Temperature, y=fire_area.Fire_Area, line_kws={"color": "red"})
plt.xlabel("Temperature (C)")
plt.ylabel("Fire Area")
plt.title("Fire Area vs Temperature (C)")
```

```python
#Graph 2: Fire Area vs Humidity

#Construct Linear Regression
humidity_linregress = stats.linregress(x=fire_area.Humidity,
                                        y=fire_area.Fire_Area)

#Print Out Assessment Statistics
print("Graph 2: ")
print(f"The slope is: {humidity_linregress.slope}")
print(f"The intercept is: {humidity_linregress.intercept}")
print(f"The r^2 value is {humidity_linregress.rvalue ** 2}")
print(f"The p-value is {humidity_linregress.pvalue}")

#Print Out Equation and Calculate Prediction
print()
print(f"The linear regression equation is y = {humidity_linregress.slope} * x + {humidity_linregress.intercept}")
humidity_prediction = humidity_linregress.slope * 95 + humidity_linregress.intercept
print(f"The predicted fire area when humidity is 95% is {humidity_prediction:.2f} units.")

#Analysis
print()
print("Given the r^2 value of approximately 0.0057, this means that the independent variable, humidity, can explain only 0.57% of the variation in the dependent variable, fire area. This indicates that other factors are more significant in assessing fire area. While humidity may still have some effect, the relationship is extremely weak. 99.43% of the variation in fire area is explained by external factors not captured by humidity.")
print()
print("With a p-value of approximately 0.0863, assuming that we are at a confidence interval of 95%, since our p-value is greater than the alpha of 0.05, humidity does not have a statistically significant relationship with fire area. We cannot confidently conclude that humidity affects fire area." )

#Construct Graph
humidity_model = sns.regplot(x=fire_area.Humidity, y=fire_area.Fire_Area, line_kws={"color": "red"})
plt.xlabel("Humidity (%)")
plt.ylabel("Fire Area")
plt.title("Fire Area vs Humidity (%)")
```

```python
#Graph 3: Fire Area vs Wind

#Construct Linear Regression
wind_linregress = stats.linregress(x=fire_area.Wind,
                                    y=fire_area.Fire_Area)

#Print out Assessment Statistics
print("Graph 3: ")
print(f"The slope is: {wind_linregress.slope}")
print(f"The intercept is: {wind_linregress.intercept}")
print(f"The r^2 value is {wind_linregress.rvalue ** 2}")
print(f"The p-value is {wind_linregress.pvalue}")

#Print Out Equation and Calculate Prediction
print()
print(f"The linear regression equation is y = {wind_linregress.slope} * x + {wind_linregress.intercept}")
wind_prediction = wind_linregress.slope * 8 + wind_linregress.intercept
print(f"The predicted fire area when wind is 8 mph is {wind_prediction:.2f} units.")

#Analysis
print()
print("Given the r^2 value of approximately 0.0002, this means that the independent variable, wind, can explain only 0.02% of the variation in the dependent variable, fire area. This indicates that other factors are more significant in assessing fire area. While wind may still have the tiniest effect, the relationship is extremely weak and we cannot factor it as a true predictor for fire area. 99.98% of the variation in fire area is explained by external factors not captured by wind.")
print()
print("With a p-value of approximately 0.7799, assuming that we are at a confidence interval of 95%, since our p-value is greater than the alpha of 0.05, wind does not have a statistically significant relationship with fire area. We cannot conclude that wind affects fire area. In fact, it seems as though there is almost no evidence that wind impacts fire area." )

#Construct Graph
wind_model = sns.regplot(x=fire_area.Wind, y=fire_area.Fire_Area, line_kws={"color": "red"})
plt.xlabel("Wind (mph)")
plt.ylabel("Fire Area")
plt.title("Fire Area vs Wind (mph)")
```
